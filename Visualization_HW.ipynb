# Задание 1.1

Изобразите график функции по заданным точкам x и y цветом 'magenta'.

Изобразите сетку синим цветом точечной линией толщиной 0.3.

Напишите "легенду" графика, подпишите оси, дайте название графику. 

Изобразите график в диапазоне x от -10 до 10  и y от -10 до 10, и пусть шкала идет для каждого второго целого значения (-10, -8, -6 ...).

```python
t = np.linspace(0, 60, 10000)
x=4.4*(np.cos(t)+np.cos(1.1*t)/1.1)
y=4.4*(np.sin(t)-np.sin(1.1*t)/1.1) 
```

```python
plt.plot(x, y, color='magenta')
plt.grid(color='b', ls='--', linewidth=0.3)
plt.axis([-10, 10, -10, 10])
plt.title('Magenta star')
ax = plt.gca()
ax.set_xlabel("X", fontsize=15, color='blue') 
ax.set_ylabel("Y", fontsize=15, color='orange')
plt.text(2, 8, "Magenta")

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.xticks(np.arange(-10,11,2))
plt.yticks(np.arange(-10,11,2))
plt.show()
```

# Задание 1.2 

Найдите еще какой-нибудь интересный график и нарисуйте его:

```python
# Import Data
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")

# Draw Plot
plt.figure(figsize=(13,10), dpi= 80)
sns.violinplot(x='class', y='hwy', data=df, scale='width', inner='quartile')

# Decoration
plt.title('Violin Plot of Highway Mileage by Vehicle Class', fontsize=22)
plt.show()
```

# Задание 2

Изобразите 4 графика (2 х 2) для функций sin, cos, tan, cot с помощью subplot. (Будьте внимательны, в некоторых точках функции могут не существовать).

Красиво оформите эти графики, сделайте все подписи, подберите размеры так, чтобы надписи разных графиков не перекрывались.

```python
x = np.linspace(-10, 10, 1000)
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x), color='magenta')
ax[1].plot(x, np.cos(x),  color='g')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('Синус')
ax[1].set_title('Косинус')

plt.subplots_adjust(wspace=1, hspace=1)
plt.show()

fig.savefig('temp.png')
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.tan(x))
ax[1].plot(x, 1/np.tan(x), color='r')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('Тангенс')
ax[1].set_title('Котангенс')
plt.subplots_adjust(wspace=1, hspace=1)
plt.show()

fig.savefig('temp.png')
```
