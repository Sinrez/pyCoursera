```python
import numpy as np
from sklearn.linear_model import LinearRegression
import scipy as sp
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import pandas as pd
import math
import operator
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
iris_set = sns.load_dataset("iris")
```

```python
axes = plt.axes()
axes.set_xlim([4, 8])         
axes.set_ylim([0, 8])          
legend = []
plt.title('Линейная регрессия - Ирисы Фишера')
x_lable='sepal_length'
y_lable='petal_length'
axes.set_xlabel(x_lable)             
axes.set_ylabel(y_lable)  
for c, rows in iris_set.groupby('species'):
    plt.scatter(rows[x_lable], rows[y_lable])
    legend.append("species=%s" % c)
plt.legend(legend, loc="upper left")
plt.grid()
plt.show()
```

```python
x_lable='sepal_length'
y_lable='petal_length'
x= iris_set[x_lable].values.reshape((-1, 1))
y= iris_set[y_lable].values
```

```python
model = LinearRegression()
model.fit(x, y)
print('coefficient of determination:', model.score(x, y))
print('intercept:', model.intercept_) # b₀
print('slope:', model.coef_) # b₁
```

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), color='red', linewidth=2);
```

```python
axes = plt.axes()
axes.set_xlim([4, 8])         
axes.set_ylim([0, 8])  
legend = []
plt.title('Линейная регрессия - Ирисы Фишера. Пример 2.')
x_lable='sepal_length'
y_lable='petal_length'
axes.set_xlabel(x_lable)             
axes.set_ylabel(y_lable)
for c, rows in iris_set.groupby('species'):
    X, Y = rows[x_lable].values.reshape((-1, 1)), rows[y_lable].values
    model.fit(X, Y)
    print(c)
    print('coefficient of determination:', model.score(X, Y))
    print('intercept:', model.intercept_) # b₀
    print('slope:', model.coef_) # b₁
    plt.scatter(X,Y)
    legend.append("species=%s" % c)
    plt.plot(X, model.predict(X), linewidth=2)
plt.legend(legend, loc="upper left")
plt.grid()
plt.show()
```
